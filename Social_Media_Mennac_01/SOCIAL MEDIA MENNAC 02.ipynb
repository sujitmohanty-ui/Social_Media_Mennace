```python
# Define the bins and labels
age_bins = [0, 18, 30, 45, 60, 100]
age_labels = ['Teenager', 'Young Adult', 'Adult', 'Middle-aged', 'Senior']

# Categorize Age into bins
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
```

```python
print(df[['Age', 'Age Group']].head())
```

```python
df.groupby('Age Group')['Number of Videos Watched'].mean().plot(kind='pie',autopct='%1.1f%%',shadow=True)
```

```python
df.groupby('Age Group')['Number of Videos Watched'].mean().idxmax()
```

```python
df.groupby(['Age Group','Platform'])['Total Time Spent'].mean()
```

```python
sns.barplot(x='Age Group',y='Total Time Spent',hue='Platform',data=df);
plt.title('Total Time Spent on Platform by Age Group and Platform Type');
```

```python
addiction_bins = [0, 2, 5, 7]
addiction_labels = ['Low', 'Moderate', 'High']

# Categorize Addiction Level into bins
df['Addiction Category'] = pd.cut(df['Addiction Level'], bins=addiction_bins, labels=addiction_labels)
```

```python
df['Addiction Category'].value_counts().plot(kind='barh',title='Distribution of Addiction Category',xlabel='counts');
```

```python
df.groupby(['Age Group','Addiction Category'])['Total Time Spent'].mean()
```

```python
sns.barplot(x='Age Group',y='Total Time Spent',hue='Addiction Category',data=df);
plt.title('Total Time Spent by Age Group and Addiction Category');
```

```python
plt.figure(figsize=(15,6))
plt.title('Scroll Rate Across Video Categories by Addiction Level')
sns.lineplot(x='Video Category',y='Scroll Rate',hue='Addiction Category',data=df);
```

```python
plt.figure(figsize=(15,6))
plt.title('Scroll Rate by Video Length Across Age Groups')
sns.lineplot(x='Video Length',y='Scroll Rate',hue='Age Group',data=df);
```
